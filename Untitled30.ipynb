```python
import numpy as np
import pandas as pd
```

```python
data =[50, 60, 65, 70, 75, 80, 85, 90]
df = pd.DataFrame(data,columns=['Mark'])
df['Mark']=df['Mark'].replace(90,900)
print(df)
means= df.mean()
print(f"Mean Value = {means}")
median = df.median()
print(f"Median Value = {median}")
mode =df.mode()
print(f"Mode Value ={mode}")
#Outlier (900) mean মানকে অনেক বেশি পরিবর্তন করে দেয়, কিন্তু Median তেমন বদলায় না।
#তাই যদি ডেটায় outlier থাকে, তাহলে Median কেন্দ্রীয় প্রবণতা (central tendency) মাপার জন্য ভালো সূচক।
```

```python
data =[2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5]
df = pd.DataFrame(data,columns=['Cat weights'])
Q1 = np.percentile(df['Cat weights'],25)
print(f"Q1 ={Q1}")
Q2 =np.percentile(df['Cat weights'],50)
print(f"Q2 ={Q2}")
Q3=np.percentile(df['Cat weights'],75)
print(f"Q3={Q3}")
# amra data valaue ke 4 vage vag korte pari Q1=25%,Q2=50%,Q3=75%,Q4=100
#kintu Q3=75% er upor gele amara outlier dhori
IQR = Q3 - Q1
print(f"IQR  is {IQR:2f}")
low = Q1-1.5*IQR
high =Q3 + 1.5*IQR
print(f"Low = {low:.2f} to")
print(f"High = {high:.2f}")
outliers = df[(df['Cat weights'] < low) | (df['Cat weights'] > high)]
if outliers.empty:
    print("No possible outliers.")
else:
    print("Possible outliers:\n", outliers)
# low er khetre ei khane 2.33 niche kono value nei ar high er khetre 4.73 upor kono
# value nei tai bola jay ei khane kono outlier nei

mu =50
sigma =10
x1=65
z = ((x1-mu)/sigma)
print(f"Z Value ={z}")
# x find
z1=-2.5
x =mu+(z1*sigma)
print(f"X value {x}")
```

```python
#1. In a survey of 100 people:
#40 like pizza (A)
#50 like burgers (B)
#20 like both (A ∩ B)
#Find P(A), P(B), P(A ∩ B), and P(A | B).
total =100
PA = 40
PB =50
PA_PB =20
P_A_find = PA/total
print(f"P(A) = {P_A_find}")
PB_find = PB/total
print(f"P(B) = {PB_find}")
P_A_PB_find = PA_PB/total
print(f"P(A ∩ B) = {P_A_PB_find}")
```

```python
# Given data
total_emails = 1000
spam = 100
spam_with_free = 40
nonspam_with_free = 60

# Probabilities
P_S = spam / total_emails
P_F_given_S = spam_with_free / spam
P_F = (spam_with_free + nonspam_with_free) / total_emails

# Bayes' theorem
P_S_given_F = (P_F_given_S * P_S) / P_F
per = P_S_given_F*100
print(f"P(S|F) = {per:1f}%")
```
