```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
hours = np.array([2,4,6,8,10]).reshape(-1,1)
marks = np.array([55,60,70,80,88])
hours,marks
```

```python
model = LinearRegression()
model.fit(hours,marks)
```

```python
print('m :',model.coef_[0])
print('c :',model.intercept_)
```

```python
plt.scatter(hours,marks,color='blue',label='Actual Data')

Pre_marks = model.predict(hours)
plt.plot(hours,Pre_marks,color='red',label="Best fit Line")
plt.legend()
plt.xlabel("Study_Hours")
plt.ylabel("Marks")
plt.title("Hours vs Marks")
plt.show()
```

```python
new_hours = np.array([[7]])
pre = model.predict(new_hours)
print("Predicted Mark: ", pre[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
hours = np.array([2, 4, 6, 8, 10]).reshape(-1, 1)
marks = np.array([55, 60, 70, 80, 88])

hours, marks
```

```python
lin_reg = LinearRegression()
lin_reg.fit(hours, marks)
```

```python
print("m :", lin_reg.coef_[0])
print("c :", lin_reg.intercept_)
```

```python
pre_marks = lin_reg.predict(hours)
mse_lin = mean_squared_error(marks, pre_marks)
print("MSE for LinearRegression :",mse_lin)
```

```python
plt.scatter(hours, marks, label="Actual Data")

plt.plot(hours, pre_marks, label="Best Fit Line (LinearRegression)")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks – Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

```python
sdg_reg = SGDRegressor(
    max_iter=1000,
    learning_rate="invscaling",
    eta0=0.01,
    random_state=42
)

sdg_reg.fit(hours,marks)

print("m :", sdg_reg.coef_[0])
print("c :",sdg_reg.intercept_[0])

pred_sdg = sdg_reg.predict(hours)
mse_sdg = mean_squared_error(marks,pred_sdg)
print("MSE for SGDRegressor :",mse_sdg)
```

```python
plt.scatter(hours, marks, label="Actual Data")

plt.plot(hours, pre_marks, label="LinearRegression", linestyle="--")
plt.plot(hours, pred_sdg, label="SGDRegression", linestyle="-.")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression vs SGDRegressor")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Create Dataset
hours = np.array([2, 4, 6, 8, 10]).reshape(-1,1)
marks = np.array([55, 60, 70, 80, 88])
hours, marks
```

```python
#Fit Linear Regression
model = LinearRegression()
model.fit(hours, marks)
pred = model.predict(hours)
print(pred)
```

```python
#Compute Metrics
mae = mean_absolute_error(marks,pred)
mse = mean_squared_error(marks, pred)
rmse = np.sqrt(mse)
r2 = r2_score(marks, pred)
```

```python
#Print Results
print("MAE: ", mae)
print("MSE", mse)
print("RMSE", rmse)
print("R^2:",r2)
```

```python
#Visualizing the Fit
plt.scatter(hours, marks, label="Actual Data")
plt.plot(hours, pred, color="red", label="Predicted Line")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Regression Line with Actual Data")
plt.legend()
plt.grid(True)
plt.show()
```
