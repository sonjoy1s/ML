```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours =[2,3,4,1,5,7,3]
day = [1,2,3,4,5,6,7]
plt.plot(day,hours)
plt.title("Student Overview Graph")
plt.xlabel("Days")
plt.ylabel("Hours")
```

```python
df = pd.read_csv('/content/enrollment_data.csv')
df
```

```python
plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.title('Enrollment Overview')
```

```python
from matplotlib.lines import lineStyles
plt.plot(df['Year'],df['Programming'],label='Programming',color='red',marker='.')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital marketing',color='orange',marker='o')
plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.title('Enrollment Overview')
plt.legend()
```
