```python
# Perform one-hot encoding on the specified categorical columns.
# pd.get_dummies automatically drops the original categorical columns and adds the new dummy variables.
# This ensures no duplication of other columns like 'Happiness_Index(1-10)'.
df_en = pd.get_dummies(df, columns=categorical_col, dtype=int)
```

```python
x = df_en.drop(columns=[target_col]) # df_en should contain numerical and one-hot encoded features
y = df_en[target_col] # copy the target column from df_en, which should now be 1D
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42, k_neighbors=3)
X_train_res, y_train_res = sm.fit_resample(x_train, y_train)
```

```python
scaller = MinMaxScaler()
x_train_scaled = scaller.fit_transform(x_train)
x_test_scaled = scaller.transform(x_test)
```

```python
x_train_std_df = pd.DataFrame(x_train_scaled,columns=x_train.columns,index=x_train.index)
x_test_std_df = pd.DataFrame(x_test_scaled,columns=x_test.columns,index=x_test.index)
display(x_train_std_df.head())
display(x_test_std_df.head())
```

```python
from sklearn.linear_model import LogisticRegression
scaller = StandardScaler()
x_train_scaled = scaller.fit_transform(X_train_res) # Use oversampled X_train_res
x_test_scaled = scaller.transform(x_test)
model = LogisticRegression()
model.fit(x_train_scaled, y_train_res) # Use oversampled y_train_res
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# y_test is now a 1D Series and y_pred is a 1D array
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
