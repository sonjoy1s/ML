```python
import matplotlib.pyplot as plt
import numpy as np
x =['jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
y =[15,18,24,29,32,33,31,30,28,26,22,17]
plt.plot(x,y, color='Green', marker='o', linestyle='--')
plt.title("Average Monthly Temperature in Daffodil City (2024)")
plt.xlabel("Month")
plt.ylabel("Temperature(C)")
plt.grid(True, linestyle='--',linewidth=0.5)
plt.xticks(rotation=45) # 45 dgree ghuriye dekhabe
plt.yticks(rotation=45)
plt.text("June",32.5,"Hottest Month",color='red')
plt.show()
```

```python
x =['jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']

tmp_a =[15, 18, 24, 29, 32, 33, 31, 30, 28, 26, 22, 17]
tmp_b =[10, 12, 18, 25, 28, 30, 29, 28, 26, 23, 19, 14]

plt.plot(x,tmp_a, color='blue',marker='o', linestyle='--',label='City_A')
plt.plot(x,tmp_b, color='green',marker='o', linestyle='--',label='City_B')
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature(C)")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True,color='gray',linestyle='--',linewidth=0.5)
plt.show()
```

```python
x =['jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
store_A =[120,135,150,160,170,180,175,165,155,145,130,125]
store_B =[100,115,130,145,155,165,160,150,140,135,120,110]
plt.plot(x,store_A, color='green',marker='o',linestyle='--',label='Store A')
plt.plot(x,store_B, color='blue',marker='o',linestyle='-',label='Store B')
plt.title("The monthly sales data for two stores in 2024:")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Sales(in units)")
plt.legend()
plt.grid(True,color='gray',linestyle='--',linewidth=0.5)
plt.show()
```

# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
x =['Jan','Feb','Mar','Apr','May','Jun']
sales =[120,135,150,160,170,190]
plt.bar(x,sales,color='green')
plt.xlabel("Month")
plt.ylabel("Sales(in units)")
plt.title("Monthly Sales Analysis")
plt.xticks(rotation=45)
#plt.barh(x,sales,color='blue')
plt.show()
```

```python
plt.barh(x, sales, color='blue') # horizontal BarChart
plt.xlabel("Sales (units)")
plt.ylabel("Month")
plt.title("Monthly Sales of Store A")
plt.show()
```

```python
store_A = [120, 135, 150, 160, 170, 180]
store_B = [100, 115, 130, 145, 155, 165]
b= np.arange(len(x))
width = 0.35
plt.bar(b-width/2,store_A,width,color='green',label='Store A')
plt.bar(b+width/2,store_B,width,color='blue',label='Store B')
plt.xlabel("Month")
plt.ylabel("Sales(in units)")
plt.title("Monthly Sales Analysis")
plt.legend()
plt.xticks(b,x)
plt.show()
```
