```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(0)

# 100 students
student_id = np.arange(1, 101)
attendance = np.random.randint(60, 101, size=100)   # 60% - 100%
study_hours = np.random.randint(1, 10, size=100)    # 1 - 9 hours/day
exam_score = attendance * 0.4 + study_hours * 6 + np.random.randn(100) * 5  # mix formula
df = pd.DataFrame({
    'Student_ID': student_id,
    'Attendance(%)': attendance,
    'Study_Hours': study_hours,
    'Exam_Score': exam_score
})

print(df.head())
print("\nSummary Statistics:")
print(df.describe())

print("\nTop 5 Students by Score:")
print(df.nlargest(5, 'Exam_Score'))

print("\nAverage Exam Score:", df['Exam_Score'].mean())
plt.plot(df['Student_ID'], df['Exam_Score'], color='green', linestyle='--', marker='o')
plt.title("Exam Score Trend by Student ID")
plt.xlabel("Student ID")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
plt.scatter(df['Study_Hours'], df['Exam_Score'], color='blue')
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours per Day")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
avg_score = df.groupby('Study_Hours')['Exam_Score'].mean()

plt.bar(avg_score.index, avg_score.values, color='orange')
plt.title("Average Exam Score by Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Average Score")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
df.to_csv("student_performance.csv", index=False)
print("✅ Dataset saved as student_performance.csv")
```
