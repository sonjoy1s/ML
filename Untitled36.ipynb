```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/train_and_test2.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#df['Sex']=df['Sex'].map({'male':0,'female':1})
```

```python
features = ['Sex','Fare','Pclass','Age', 'sibsp', 'Parch', 'Embarked']
target = ['2urvived']
deleted_features = ['zero','zero.1','zero.2','zero.3','zero.4',
                    'zero.5','zero.6','zero.7','zero.8','zero.9','zero.10','zero.11','zero.12',
                    'zero.13','zero.14','zero.15','zero.16']
```

```python
df.drop(columns=deleted_features,inplace=True, errors='ignore')
```

```python
corr_matrix = df[features + target].corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
x = df[features]
y_label = df[target]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_label, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
y_pred =model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
output = pd.DataFrame({'PassengerId': df.loc[x_test.index, 'Passengerid'], 'Survived': y_pred})
output.to_csv('my_submission.csv',index=False)
```
