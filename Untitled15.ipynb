```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('/content/Sales Data.csv.zip', compression='zip')
df.head()
df.info()
df.describe()
# missing value handle
#df.isnull().sum()
#df.dropna(inplace=True) # removed row with missing value
```

```python
#created a new col
df['Total Sales'] = df['Quantity Ordered']*df['Price Each']
df['Oder Date']= pd.to_datetime(df['Order Date'])
df['Month']= df['Oder Date'].dt.month
df['Hour']=df['Oder Date'].dt.hour
df['City']=df['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
```

```python
monthly_sales = df.groupby('Month')['Total Sales'].sum()
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
top_products = df.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='barh', color='lightgreen')
plt.title('Top 5 Products by Quantity Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Product')
plt.show()
```

```python
city_sales = df.groupby('City')['Total Sales'].sum().sort_values(ascending=False)
city_sales.plot(kind='bar', color='salmon')
plt.title('Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
hourly_orders = df.groupby('Hour')['Order ID'].count()
hourly_orders.plot(kind='line', marker='o', color='purple')
plt.title('Hourly Order Trend')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()
```
