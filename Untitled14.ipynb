```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

```python
age = np.array(ages)
print(age)
print(age.dtype)
print(age.ndim)
print(age.shape)
print(age.size)

print("salary part :")

salary = np.array(salaries)
print(salary)
print(salary.dtype)
print(salary.ndim)
print(salary.shape)
print(salary.size)
```

```python
salary_max = np.max(salary)
print(salary_max)

salary_min = np.min(salary)
print(salary_min)

salary_avg = np.mean(salary)
print(salary_avg)
```

```python
filter_age = age[age>30]
print(filter_age)

len_age = len(filter_age)
print(len_age)
```

```python
incr_age = age + 5
print(incr_age)
```

```python
salary_sum = np.sum(salary)
print(salary_sum)

diff_salary = salary_max - salary_min
print(diff_salary)
```
